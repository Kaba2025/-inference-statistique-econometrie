```python
import pandas as pd

# Charger le fichier CSV
file_path = r"C:\INSEEDS\PROJET\STAT DESC\Student_Depression.csv"
df = pd.read_csv(file_path)

# Aperçu des premières lignes
df.head()
```

```python
# Nombre d'observations (lignes)
nombre_observations = df.shape[0]
print(f"Nombre d'observations : {nombre_observations}")
```

```python


# Afficher le nombre de valeurs manquantes par colonne
missing_values = df.isnull().sum()
print("Nombre de valeurs manquantes par colonne :\n")
print(missing_values)

# Total des valeurs manquantes
total_missing = missing_values.sum()
print(f"\nNombre total de valeurs manquantes dans le dataset : {total_missing}")
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Visualisation des valeurs manquantes
plt.figure(figsize=(12,6))
sns.heatmap(df.isnull(), cbar=False, cmap='viridis')
plt.title("Valeurs manquantes - Avant traitement")
plt.show()

# Suppression ou imputation si <5%
df = df.dropna(thresh=df.shape[0]*0.95, axis=1)  # on garde si >=95% de valeurs présentes
df = df.dropna()  # ou bien .fillna(...) selon la logique métier

# Visualisation après traitement
plt.figure(figsize=(12,6))
sns.heatmap(df.isnull(), cbar=False, cmap='viridis')
plt.title("Valeurs manquantes - Après traitement")
plt.show()
```

```python
# Supprimer les doublons
df = df.drop_duplicates()
```

```python
from scipy.stats.mstats import winsorize

# Variables quantitatives
quant_vars = df.select_dtypes(include=['float64', 'int64']).columns

# Boxplot avant traitement
df[quant_vars].plot(kind='box', subplots=True, layout=(len(quant_vars)//3+1, 3), figsize=(15,10), sharex=False)
plt.suptitle("Boxplot avant winzorisation")
plt.tight_layout()
plt.show()

# Winzorisation (à 5%)
for var in quant_vars:
    df[var] = winsorize(df[var], limits=[0.05, 0.05])

# Boxplot après traitement
df[quant_vars].plot(kind='box', subplots=True, layout=(len(quant_vars)//3+1, 3), figsize=(15,10), sharex=False)
plt.suptitle("Boxplot après winzorisation")
plt.tight_layout()
plt.show()
```

```python
import scipy.stats as stats

# Histogrammes
df[quant_vars].hist(bins=15, figsize=(15,10), edgecolor='black')
plt.suptitle("Distribution des variables quantitatives")
plt.tight_layout()
plt.show()

# Résumé statistique complet
summary_stats = df[quant_vars].agg(['mean', 'median', 'min', 'max', 'std', 'var', stats.kurtosis, stats.skew]).T
summary_stats.columns = ['Moyenne', 'Médiane', 'Min', 'Max', 'Écart-Type', 'Variance', 'Kurtosis', 'Asymétrie']
summary_stats
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import kurtosis, skew



# Sélection des variables quantitatives (automatique ou manuelle si besoin)
quanti_vars = df.select_dtypes(include=['float64', 'int64']).columns.tolist()
quanti_vars.remove("id")  # On enlève l'identifiant

print("Variables quantitatives :", quanti_vars)
```

```python
# Taille du graphique
plt.figure(figsize=(16, 12))

# Afficher les distributions
for i, col in enumerate(quanti_vars):
    plt.subplot(3, 3, i + 1)
    sns.histplot(df[col], kde=True, bins=30, color='skyblue')
    plt.title(f'Distribution de {col}')
    plt.xlabel('')
    plt.ylabel('')

plt.tight_layout()
plt.show()
```

```python
# Fonction pour calculer les stats demandées
def statistiques_perso(df, variables):
    data = []
    for var in variables:
        stats = {
            'Variable': var,
            'Moyenne': df[var].mean(),
            'Médiane': df[var].median(),
            'Min': df[var].min(),
            'Max': df[var].max(),
            'Écart-type': df[var].std(),
            'Variance': df[var].var(),
            'Asymétrie (Skew)': skew(df[var].dropna()),
            'Kurtosis': kurtosis(df[var].dropna())
        }
        data.append(stats)
    return pd.DataFrame(data)

# Résumé des stats
resume_stats = statistiques_perso(df, quanti_vars)
print(resume_stats)
```

```python
# Variables qualitatives
qual_vars = df.select_dtypes(include=['object']).columns

for var in qual_vars:
    fig, axs = plt.subplots(1, 2, figsize=(10,5))
    
    df[var].value_counts().plot(kind='bar', ax=axs[0], color='skyblue', edgecolor='black')
    axs[0].set_title(f"Diagramme en bande - {var}")
    
    df[var].value_counts().plot(kind='pie', autopct='%1.1f%%', ax=axs[1])
    axs[1].set_title(f"Diagramme en secteur - {var}")
    axs[1].set_ylabel("")
    
    plt.tight_layout()
    plt.show()

# Tableaux d’effectifs et fréquences
for var in qual_vars:
    tab = df[var].value_counts().to_frame('Effectif')
    tab['Fréquence (%)'] = round(tab['Effectif'] / tab['Effectif'].sum() * 100, 2)
    print(f"\nTableau pour {var} :\n", tab)
```
